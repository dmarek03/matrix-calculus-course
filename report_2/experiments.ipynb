```python
from utils.matrix_generator import generate_matrix
from algortihms.gauss_elimination_with_pivot import gauss_with_pivot
from algortihms.lu_factorization import lu_decomposition
```

```python
A = generate_matrix(3)
b = generate_matrix(3, True)
print("A:")
print(A)
print("b:")
print(b)
```

```python
X=gauss_with_pivot(A, b)
```

```python
print("Weryfikacja Ax =", A @ X)
print("Weryfikacja b =", b)
```

```python
L, U, x = lu_decomposition(A, b)

print("Macierz L:")
print(L)
print("\nMacierz U:")
print(U)
print("\nWeryfikacja L @ U == A:", (L @ U, A))
print("\nRozwiązanie x:", x)
print("Weryfikacja A @ x == b:", (A @ x, b))
```
