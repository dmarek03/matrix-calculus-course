```python
from utils.matrix_generator import generate_matrix
from algortihms.gauss_elimination_with_pivot import gauss_with_pivot
from algortihms.gauss_elimination import gauss_no_pivot
from algortihms.solve_lus import solve_lu
from report_2.benchmark import benchmark
```

```python
A = generate_matrix(5)
b = generate_matrix(5, True)
print("A:")
print(A)
print("b:")
print(b)
```

```python
x1 = gauss_with_pivot(A, b)
print("Rozwiązanie (Gauss z pivotem):")
print(x1)
x2 = gauss_no_pivot(A, b)
print("Rozwiązanie (Gauss bez pivotu):")
print(x2)
```

```python
X_lu_pivot = solve_lu(A, b, algorithm="lu_with_pivot")
print("Rozwiązanie (LU z pivotem):")
print(X_lu_pivot)
print("Sprawdzenie (A @ X_lu_pivot):")
print(A @ X_lu_pivot)
```

```python
x_no_pivot = solve_lu(A, b, algorithm="lu_no_pivot")
print("Rozwiązanie (LU bez pivotu):")
print(x_no_pivot)
print("Sprawdzenie (A @ x_no_pivot):")
print(A @ x_no_pivot)
```
