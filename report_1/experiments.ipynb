```python
from benchmark import benchmark
from utils.plots import plot_results_multi
```

```python
matrix_sizes = [2**i for i in range(2, 9)]
l_parameters = [2**3, 2**4, 2**5, 2**6, 2**7]

results = {}

for idx, l_parameter in enumerate(l_parameters):
    times, ops_counts = benchmark(sizes=matrix_sizes, repeats=3, l_size=l_parameter)
    results[f"l={l_parameter}"] = (times, ops_counts)
    print(
        f"Progress: {(idx + 1) / len(l_parameters) * 100:.1f}% (l={l_parameter}) (times={times})"
    )

plot_results_multi(matrix_sizes, results)
```
